#### Continuous compared to Discontinuous Functions
*  Limits are easily calculated for a continuous function
*  For example, find the limit as $x$ approaches 5 in the expression:
\begin{equation*}
\lim_ {x \to 5} x^2 + 2x + 2
\end{equation*}

```python
x = np.linspace(-4, 10, 100) # start, finish, number of points
```

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-4, 10])
ax.set_ylim([-5, 90])
plt.plot(x, y, color='b');
plt.axhline(y= 5**2 + 2*5 + 2, linestyle='--', color='purple');
plt.axvline(x=5, linestyle='--', color='purple');
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\lim_{x \\to 5} \\hspace{1} x^2 + 2x + 2$")
plt.grid()
plt.tight_layout()
plt.show()
```

*  Here we can just substitute 5 for $x$ in the equation because this is a continuous function.
\begin{equation*}
\lim_{x \to \color{purple}{5}}\hspace{0.1cm}\color{purple}{x}^2 + 2\color{purple}{x} + 2
=
\color{purple}{5}^2 + 2\times{\color{purple}{5}} + 2
= 25 + 10 + 2
=
37
\end{equation*}


*  However, some functions are not continuous. For example:
\begin{equation*}
\lim_{x \to 1} \frac{x^2-1}{x-1}
\end{equation*}
*  Although this plot appears to be continuous, this is a discontinuous function because at $x=1$ there is no value and the function actually skips an infinitesily small point at $x=1$. We can calculate values very close to $x=1$ from the right or the left but at $x=1$ there is a divide by zero situation so the function does not exist at $x=1$.

```python
x = np.linspace(-1, 5, 5)
```

```python
y = (x**2 -1) / (x - 1)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1, 5])
ax.set_ylim([-1, 5])
y[x==1] = np.nan # replace the value at x=1 with nan
plt.plot(1, 2, marker='o', fillstyle='none')
plt.plot(x, y, color='b');
plt.axhline(y=2, linestyle='--', color='purple');
plt.axvline(x=1, linestyle='--', color='purple');
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\lim_{x \\to 1}\\frac{x^2-1}{x-1}$")
plt.grid()
plt.tight_layout()
plt.show()
```

*  We can explore how getting closer and closer to 1 from the left and right approaches 2 but if we let $x=1$ we get a divide by zero error.

```python
def discon_func(x_val):
  y_val = (x_val**2 - 1)/(x_val - 1)
  return y_val
```

```python
# If we set x = 1 then we get a divide by zero error
#discon_func(1)
```

```python
# We can observe that we get very close to 2 as we enter values near but not at 1.
print(discon_func(0.9))
print(discon_func(0.999))
print(discon_func(1.001))
print(discon_func(1.1))
```

##### Solving Limits using Factoring
*  Sometimes it is possible to solve for the limit using the algebraic method of factoring. For example:
\begin{equation*}
\lim_{x \to 1}\frac{x^2 - 1}{x - 1} = \lim_{x \to 1}\frac{(x - 1)(x + 1)}{(x - 1)}=\lim_{x \to 1}\frac{(x - 1)(x + 1)}{(x - 1)} = \lim_{x \to 1}(x + 1) = 1 + 1 = 2
\end{equation*}

##### Solving Limits using Approaching Method
*  Sometimes factoring does not work either. For example, consider the following function:
\begin{equation*}
\lim_{x \to 0}\frac{\sin{x}}{x}
\end{equation*}
*  Here we have a function with a divide by zero condition that cannot be factored to remove the $x$ from the denominator. To solve this we can use an appraoch method to solve the limit.

```python
x = np.linspace(-10, 10, 1000) # start, finish, number of points
y = np.sin(x)/x

fig, ax = plt.subplots()
ax.set_xlim([-10, 10])
ax.set_ylim([-1, 2])
y[x==0]=np.nan # replace the value at x=0 with np.nan
plt.plot(0, 1, marker='o', fillstyle='none')
plt.plot(x, y, color='b');
plt.axhline(y=1, linestyle='--', color='purple');
plt.axvline(x=0, linestyle='--', color='purple');
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\lim_{x \\to 0}\\hspace{1}\\frac{\\sin{x}}{x}$")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
def sin_func(x_val):
  y_val = np.sin(x_val) / x_val
  return y_val
```

```python
# If we input a 0 as the x value we get a divide by zero error.
y = sin_func(0)
```

```python
# But if we approach 0 from the left and the right we get very close to 1.
print(sin_func(0.1))
print(sin_func(0.001))
print(sin_func(-0.001))
print(sin_func(-0.1))
```

##### Approaching Infinity
*  There are also instances where we need to calculate the limit of a non-discrete value such as plus or minus infinity($\pm ∞$).

```python
x1 = np.linspace(-10, -0.001, 100) # start, finish, number of points
x2 = np.linspace(0.001, 10, 100)
y1 = 25/x1
y2 = 25/x2

fig, ax = plt.subplots()
ax.set_xlim([-10, 10])
ax.set_ylim([-300, 300])
#plt.plot(0, 0, marker='o', fillstyle='none')
plt.plot(x1, y1, color='b');
plt.plot(x2, y2, color='b');
#plt.axhline(y=0, linestyle='--', color='purple');
#plt.axvline(x=0, linestyle='--', color='purple');
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\lim_{x\\to\\infty}\\hspace{1}\\frac{25}{x}$")
plt.grid()
plt.tight_layout()
plt.show()
```

*  Let us consider the function plotted above and the code below. We can clearly see that as x approaches infinity in both the positive and negative direction f(x) gets closer and closer to 0. Therefore,
\begin{equation*}
\lim_{x \to \pm\infty} \frac{25}{x} = 0
\end{equation*}

```python
def inf_func(x_val):
  y_val = 25 / x_val
  return y_val
```

```python
# We can input some large x values in the positive direction to see where the function goes as we head into positive infinity.
print(inf_func(1e3))
print(inf_func(1e4))
print(inf_func(1e6))
# We can also input some extremely negative x values to see where the function goes in negative infinity.
print(inf_func(-1e3))
print(inf_func(-1e4))
print(inf_func(-1e6))
```

#### Differential Calculus
*  The Delta Method \\
\
\begin{equation*}
\frac{dy}{dx} = \lim_{Δx \to 0}{\frac{f(x + Δx) - f(x)}{Δx}}
\end{equation*} \\


*  As an example we can look at the continuous curve defined by: \\
\
\begin{equation*}
f(x) = x^2 + 2x + 2
\end{equation*}


```python
# Define a function that calculates f(x)
def f(x):
  y = x**2 + 2*x + 2
  return y
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, number of points
y = f(x)
fig, ax = plt.subplots()
ax.set_xlim([-10, 10])
ax.set_ylim([0, 100])
plt.plot(x, y, color='b');
#plt.axhline(y= 5**2 + 2*5 + 2, linestyle='--', color='purple');
#plt.axvline(x=5, linestyle='--', color='purple');
plt.xlabel("x")
plt.ylabel("y")
plt.title("$f(x) = x^2 + 2x + 2$")
plt.grid(color='lightgray', linestyle='--', linewidth=1)
plt.tight_layout()
plt.show()
```

*  By using the delta method we have derived the definition of differentiation from first principles. So the derivative of $y$(denoted as $dy$) with respect to $x$(denoted as $dx$) is represented as:
\begin{equation*}
\frac{dy}{dx} = \lim_{Δx \to 0} \frac{Δy}{Δx}
\end{equation*}


*  Therefore, $Δy = y_2 - y_1$ and thus:


\begin{equation*}
\frac{dy}{dx} = \lim_{Δx \to 0} \frac{y_2 - y_1}{Δx}
\end{equation*}


*  Now if we replace $y_1$ with $f(x)$ and $y_2$ with $f(x + Δx)$, we get the common representation of differentiation:

\begin{equation*}
\frac{dy}{dx} = \lim_{Δx \to 0}{\frac{f(x + Δx) - f(x)}{Δx}}
\end{equation*}



*  We can write some code to demonstrate the differentiation equation:

```python
# Define a function that calculates f(x)
def f(x):
  y = x**2 + 2*x + 2
  return y
```

```python
# Create a differentiation function
def dydx(f, x, d_x): # pass in a function, an x coordinate, and a delta x
  return((f(x + d_x) - f(x)) / d_x)
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
  print(dydx(f, 2, delta)) # Approaches 6
```

```python
for delta in deltas:
  print(dydx(f, -1, delta)) # Approaches 0
```

##### Derivative Notation
*  Differentiation Operator:
\begin{equation*}
\frac{d}{dx} = D
\end{equation*}
*  The first derivative of $y$ w.r.t $x$:
\begin{equation*}
\dot{y} = y' = f'(x) = \frac{dy}{dx} = \frac{d}{dx}f(x) = Df(x)
\end{equation*}
*  The second derivative of $y$ w.r.t $x$:
\begin{equation*}
\ddot{y} = y'' = f''(x) = f'(f'(x)) =  \frac{d^2y}{dx^2} = \frac{d^2}{dx^2}f(x)
\end{equation*}

##### Taking the derivative of a constant
*  If $\color{orange}{c}$ is a constant:
\begin{equation*}
\frac{d}{d\color{purple}{x}}\color{orange}{c} = 0
\end{equation*}
*  A constant has no change or variation and the slope will always be 0.


##### The Power Rule
\begin{equation*}
\frac{d}{d\color{purple}{x}}\color{purple}{x}^\color{green}{n} = \color{green}{n}\color{purple}{x}^{\color{green}{n} - 1}
\end{equation*}
* For example:
\begin{equation*}
\frac{d}{d\color{purple}{x}}\color{purple}{x}^\color{green}{4} = \color{green}{4}\color{purple}{x}^{\color{green}{4} - 1} = 4x^3
\end{equation*}


##### The Constant Product Rule
\begin{equation*}
\frac{d}{d\color{purple}{x}}(\color{orange}{c}\color{blue}{y}) = \color{orange}{c}\frac{d}{d\color{purple}{x}}(\color{blue}{y}) = \color{orange}{c}\frac{d\color{blue}{y}}{d\color{purple}{x}}
\end{equation*}
*  For example, let $\color{blue}{y} = \color{purple}{x}^4$. If we want to calculate the derivative of $2\color{blue}{y}$ then:
\begin{equation*}
\frac{d\color{blue}{y}}{d\color{purple}{x}}= 4\color{purple}{x}^3
\end{equation*} \\
\begin{equation*}
\frac{d}{d\color{purple}{x}}\color{orange}{2}\color{blue}{y} = \color{orange}{2}\frac{d\color{blue}{y}}{d\color{purple}{x}}=\color{orange}{2}(4\color{purple}{x}^3) = 8\color{purple}{x}^3
\end{equation*}


##### The Sum Rule
\begin{equation*}
\frac{d(\color{blue}{y}+\color{green}{w})}{d{x}}= \frac{d\color{blue}{y}}{dx} + \frac{d\color{green}{w}}{d{x}}
\end{equation*} \\
*  For example, let $\color{blue}{y} = \color{blue}{x}^4$ and $\color{green}{w}  = \color{green}{x}^9$, then to calculate the derivative of the sum of these two terms:
\begin{equation*}
\frac{d(\color{blue}{y}+\color{green}{w})}{d{x}} = \frac{d\color{blue}{y}}{d{x}} + \frac{d\color{green}{w}}{d{x}}=\frac{d}{d{x}}(\color{blue}{x}^4) + \frac{d}{d{x}}(\color{green}{x}^9)= 4x^3 + 9x^8
\end{equation*}

#### The Product Rule
*  The product rule is denoted as:
\begin{equation*}
\frac{d(\color{green}{w}\color{orange}{z})}{dx} = \color{green}{w}\frac{d\color{orange}{z}}{dx} + \color{orange}{z}\frac{d\color{green}{w}}{dx}
\end{equation*}
*  The product rule is useful during calculations involving several rules and when the product of $w$ and $z$ cannot be differentiated before splitting them into trwo separate terms.
* For example, let $\color{blue}{y} = (6x^3)(7x^4)$, then $\color{green}{w} = 6x^3$ and $\color{orange}{z} = 7x^4$. We can use the product rule to differentiate $\color{blue}{y}$. Of course we can just multuply the terms of $\color{blue}{y}$ and then use the power rule, but for the sake of this example we will consider the two terms separately.
* If we use the product rule then:

\begin{equation*}
\frac{d\color{green}{w}}{dx} = 18x^2 \text{, and } \frac{d\color{orange}{z}}{dx} = 28x^3
\end{equation*} \\
\begin{equation*}
\frac{d\color{blue}{y}}{dx} = \frac{d(\color{green}{w}\color{orange}{z})}{dx} = \color{green}{w}\frac{d\color{orange}{z}}{dx} + \color{orange}{z}\frac{d\color{green}{w}}{dx} = (6x^3)(28^x3) + (7x^4)(18x^2) = 168x^6 + 126x^6 = 294x^6
\end{equation*}
